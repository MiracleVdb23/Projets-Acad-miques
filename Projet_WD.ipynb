```python
def extract_surf_report(url, user_agent):
    
        
        headers = {'User-Agent': user_agent}
        page = requests.get(url, headers=headers) #extraire du texte à partir de l'URL
        soup = BeautifulSoup(page.content, 'html.parser')
        

        # Supposons que vous ayez déjà récupéré les éléments de date du site Web
        date_elements = soup.find_all('div', class_='title')


        # Supposons que vous ayez déjà récupéré les éléments de date du site Web
        # date_elements = ...

        dates = [element.b.text.strip() for element in date_elements if element.b and not element.a]

                        # Fonction pour ajuster l'année en fonction du contexte
            # Fonction pour ajuster l'année en fonction du contexte
        def adjust_year(current_date, previous_date):
            current_month = current_date.month
            previous_month = previous_date.month

            # Si le mois de la date actuelle est inférieur au mois de la date précédente,
            # et que la date précédente n'est pas le 31 décembre, ajuster l'année
            if current_month < previous_month and (previous_date.month == 12 and previous_date.day == 31):
                return current_date.replace(year=previous_date.year + 1)
            else:
                # Si l'année a déjà changé et que le mois actuel est janvier, maintenir l'année
                if current_month == 1:
                    return current_date.replace(year=previous_date.year)
                else:
                    return current_date

        # Convertir les dates en objets datetime en ajustant l'année
        formatted_dates = []
        previous_date = None
        for date in dates:
            current_date = dateparser.parse(date, languages=['fr']).replace(hour=0, minute=0, second=0, microsecond=0)
            if previous_date:
                current_date = adjust_year(current_date, previous_date)
            formatted_dates.append(current_date.strftime("%Y-%m-%d"))
            previous_date = current_date

        # Répéter les dates pour chaque élément de la journée
        days = np.repeat(formatted_dates, 8)




        
        
        #  Differentes heures
        time_elements = soup.find_all('div', class_='cell date with-border')
        times = [time.text.strip() for time in time_elements]
        
        # Différentes intervalles de la hauteur de la vague
        desc_list_vague = soup.find_all('div', class_ ="cell large waves with-border") # va chercher les informations pour lesquelles on a une balise td avec la classe desc
        desc_list_vague
        
        ### Extract the text
        desc_content_vague = []
        for i in desc_list_vague : 
            desc_content_vague.append(i.text)
        desc_content_vague
        
        # Qualité de la vague
        desc_list_star = soup.find_all('div', class_ ="cell stars with-border") # va chercher les informations pour lesquelles on a une balise td avec la classe desc
        desc_list_star
        
        star_number = []
        for i in desc_list_star:
            a = 3 - (len(str(i).split('fa fa-star-o')) - 1) 
            star_number.append(a)
        star_number
        
        
        #extraction de la vitesse et la direction du vent :
        
        desc_list = soup.find_all('div', class_ ="cell large-bis-bis with-border") 
        orientation_vent = []
        vitesse_vent=[]
        
        

        for i in desc_list:
            i.text.replace('\n','')  
            orientation= str(i).split('alt="')[1].split('"')[0]
            i.text.replace('\n','')
            vitesse_vent.append( i.text.replace('\n',''))
            orientation_vent.append(orientation)
        
        # Constitution de la dataframe
        all_extract = {}
        all_extract['days'] = days
        all_extract['times'] = times
        all_extract['star'] = star_number
        all_extract['vague'] = desc_content_vague
        all_extract['orientation_vent'] = orientation_vent
        all_extract['vitesse_vent'] = vitesse_vent

        df_extract = pd.DataFrame(all_extract)
        
                # Division de la colonne "vagues" en deux colonnes en fonction du séparateur " - "
        df_extract[['vague_min', 'vague_max']] = df_extract['vague'].str.split(" - ", expand=True)

        # Conversion des colonnes "vague_min" et "vague_max" en types numériques
        df_extract['vague_min'] = pd.to_numeric(df_extract['vague_min'], errors='coerce')
        df_extract['vague_max'] = pd.to_numeric(df_extract['vague_max'], errors='coerce')

        
                # Exporter la base de données sous forme CSV
        df_extract.to_csv("surf_data.csv")

        
        
        
        
        
        
```

```python
url= "https://www.surf-report.com/meteo-surf/lacanau-s1043.html"
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
extract_surf_report(url, user_agent)
```
